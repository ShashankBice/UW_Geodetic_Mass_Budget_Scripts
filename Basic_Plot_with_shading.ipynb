```python
%matplotlib inline
import os
import pickle

import matplotlib.pyplot as plt
```

```python
df = pickle.load(open('Standard_Errod_dh.p', 'r'))
```

```python
df.head()
```

### Plot beutified by "DON" 

```python
fig, ax= plt.subplots(figsize=(20,10), sharex=True)
ax1 = ax
#ax2 = ax[1]
leg1, = ax1.plot(df.index, df['meanGlacier'], color='black', label='meanGlacier')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.fill_between(df.index, df['E_negative'], df['E_positve'],alpha=0.5, facecolor='#0000FF')
ax1.set_ylabel('Elevation Difference (m)', color='k', fontsize=16)
ax1.tick_params('y', colors='k', labelsize='large')
ax1.set_xlabel('Elevation (m a.s.l.)', fontsize=20)
ax1.tick_params('x', colors='k', labelsize='large')

ax2 = ax1.twinx()
leg2 = ax2.bar(df.index, df['count_nonzeroGlacier'], width=25, alpha=0.3, label='count_nonzeroGlacier')
ax2.set_ylabel('Count', color='k', fontsize=16)
ax2.tick_params('y', colors='k', labelsize='large')
#ax2.tick_params('x', colors='k', labelsize='large')
ax2.set_xlabel('Elevation (m a.s.l.)', fontsize=20)

# LEGEND
plt.legend(handles=[leg1, leg2], loc='upper left', fontsize='xx-large')

# TITLE
fig.suptitle('Shashank Plot', fontsize=30, y=1.05)
fig.tight_layout()
plt.show()
```
